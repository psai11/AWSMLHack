```python
import sys
from os.path import join
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error, mean_absolute_error
from numpy import sqrt
import numpy as np
from datetime import datetime, timedelta

from prophet import Prophet
from prophet.diagnostics import performance_metrics, cross_validation
```

```python
import pandas as pd

train_data0 = pd.read_csv('dataset/train_clean_c0.csv')
train_data0.head()
```

```python
train_data1 = pd.read_csv('dataset/train_clean_c1.csv')
train_data1.head()
```

```python
city0_ids = train_data0[['ID']]
city1_ids = train_data1[['ID']]
print(city0_ids.head()) 
print(city1_ids.head())

temperatureModelxY_c0 = train_data0[['Datetime', 'Temperature']]
temperatureModelxY_c1 = train_data1[['Datetime', 'Temperature']]
print(temperatureModelxY_c0.shape, temperatureModelxY_c0.head())
print(temperatureModelxY_c1.shape, temperatureModelxY_c1.head())

train2d_c0 = train_data0.drop(columns=['ID', 'Temperature', 'Datetime'])
train2d_c1 = train_data1.drop(columns=['ID', 'Temperature', 'Datetime'])

print(train2d_c0.shape)
print(train2d_c1.shape)
```

```python
prophetPd_c0 = pd.concat([temperatureModelxY_c0, train2d_c0], axis=1)
prophetPd_c1 = pd.concat([temperatureModelxY_c1, train2d_c1], axis=1)
prophetPd_c0.columns = ['ds', 'y', 'v0','v1','v2','v3','v4','v5','v6','v7','v8','v9','v10']
prophetPd_c1.columns = ['ds', 'y', 'v0','v1','v2','v3','v4','v5','v6','v7','v8','v9','v10']
print(prophetPd_c0.shape, prophetPd_c1.shape)
prophetPd_c1
```

```python

trainProphet_c0 = prophetPd_c0.iloc[0:26266]
print(trainProphet_c0.shape)
print(trainProphet_c0.head())

testProphet_c0 = prophetPd_c0.iloc[26266:]
testProphet_c0.reset_index(inplace=True)
testProphet_c0.drop('index', axis = 1, inplace=True)
print(testProphet_c0.shape)
print(testProphet_c0.head())
```

```python

trainProphet_c1 = prophetPd_c1.iloc[0:26266]
print(trainProphet_c1.shape)
print(trainProphet_c1.head())

testProphet_c1 = prophetPd_c1.iloc[26266:]
testProphet_c1.reset_index(inplace=True)
testProphet_c1.drop('index', axis = 1, inplace=True)
print(testProphet_c1.shape)
print(testProphet_c1.head())
```

```python
model_c0 = Prophet(changepoint_prior_scale=0.01,
               seasonality_prior_scale = 10,
               yearly_seasonality=True) 

model_c0.add_regressor('v0')
model_c0.add_regressor('v1')
model_c0.add_regressor('v2')
model_c0.add_regressor('v3')
model_c0.add_regressor('v4')
model_c0.add_regressor('v5')
model_c0.add_regressor('v6')
model_c0.add_regressor('v7')
model_c0.add_regressor('v8')
model_c0.add_regressor('v9')
model_c0.add_regressor('v10')

model_c0.fit(trainProphet_c0) # fit the model with the dataframe
print('Prophet0 is trained')

model_c1 = Prophet(changepoint_prior_scale=0.01,
               seasonality_prior_scale = 10,
               yearly_seasonality=True) 

model_c1.add_regressor('v0')
model_c1.add_regressor('v1')
model_c1.add_regressor('v2')
model_c1.add_regressor('v3')
model_c1.add_regressor('v4')
model_c1.add_regressor('v5')
model_c1.add_regressor('v6')
model_c1.add_regressor('v7')
model_c1.add_regressor('v8')
model_c1.add_regressor('v9')
model_c1.add_regressor('v10')

model_c1.fit(trainProphet_c1) # fit the model with the dataframe
print('Prophet1 is trained')
```

```python
future_c0 = model_c0.make_future_dataframe(periods = testProphet_c0.shape[0], freq = '1H')  
forecast_c0 = model_c0.predict(testProphet_c0)
print(forecast_c0[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].shape)

future_c1 = model_c1.make_future_dataframe(periods = testProphet_c1.shape[0], freq = '1H')  
forecast_c1 = model_c1.predict(testProphet_c1)
print(forecast_c1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].shape)
```

```python
dataList_c0 = list(testProphet_c0.y.values)
print(len(dataList_c0))
print(len(forecast_c0['yhat']))

dataList_c1 = list(testProphet_c1.y.values)
print(len(dataList_c1))
print(len(forecast_c1['yhat']))
```

```python
rmse_c0 = sqrt(mean_squared_error(dataList_c0,forecast_c0['yhat']))
print('Prophet RMSE c0: %.3f' % rmse_c0)

mae_c0= mean_absolute_error(dataList_c0,forecast_c0['yhat'])
print('Prophet MAE c0: %.3f' % mae_c0)

rmse_c1 = sqrt(mean_squared_error(dataList_c1,forecast_c1['yhat']))
print('Prophet RMSE c1: %.3f' % rmse_c1)

mae_c1= mean_absolute_error(dataList_c1,forecast_c1['yhat'])
print('Prophet MAE c1: %.3f' % mae_c1)
```
